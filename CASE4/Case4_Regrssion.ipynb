```python
from sklearn import linear_model
import pandas as pd
```

```python
# read excel to build a dataframe
df = pd.read_excel("~/Documents/zf/SIT/BIA686/CASE4/ForAnalysis.xlsx")
df.head()
```

```python
# define x, y as independent and dependent variables
X = df[["Value_of_savings", "prev_credits"]]
Y = df["Amount_of_credits"]
```

```python
# initalize the linear model and fit it into X and Y
lm = linear_model.LinearRegression()
model = lm.fit(X,Y)
```

```python
# predict the Y based on the generated model
predictions = lm.predict(X)
df_res = pd.DataFrame(predictions, columns = ["predict_credits"])
```

```python
# sort predict_credicts for first 50 customers in descending order
df_res.sort_values(by = "predict_credits", ascending = False)[0:49]
```
